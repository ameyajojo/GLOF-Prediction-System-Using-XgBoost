```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
import glob  # Add this import at the top with other imports

# Load multiple datasets
csv_files = glob.glob("*.csv")  # This will find all CSV files in the current directory
dfs = []  # List to store individual DataFrames

for file_path in csv_files:
    try:
        temp_df = pd.read_csv(file_path, encoding="ISO-8859-1")
        dfs.append(temp_df)
        print(f"Successfully loaded: {file_path}")
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

# Combine all DataFrames into one
if dfs:
    df = pd.concat(dfs, ignore_index=True)
    print(f"Total records in combined dataset: {len(df)}")
else:
    raise ValueError("No CSV files were successfully loaded")

# Select relevant features
selected_features = [
    'Lat_lake', 'Lon_lake', 'Elev_lake',  # Location-related
    'Driver_lake', 'Driver_GLOF', 'Mechanism',  # Event properties
    'Impact_type', 'Repeat',  # Impact-related
    'Region_RGI', 'Region_HiMAP'  # Geographical information
]
df = df[selected_features + ['Impact']].copy()

# Drop rows with missing target values
df = df.dropna(subset=['Impact'])

# Define a function to classify impact into categories
def classify_impact(impact):
    if isinstance(impact, str):
        impact_lower = impact.lower()
        if "death" in impact_lower or "destroyed" in impact_lower or "severe" in impact_lower:
            return "High"
        elif "damaged" in impact_lower or "affected" in impact_lower or "displacement" in impact_lower:
            return "Medium"
        else:
            return "Low"
    return "Low"

# Apply classification to Impact column
df["Impact_Category"] = df["Impact"].apply(classify_impact)
df = df.drop(columns=["Impact"])

# Encode categorical variables
label_encoders = {}
categorical_columns = ['Driver_lake', 'Driver_GLOF', 'Mechanism', 'Impact_type', 'Repeat', 'Region_RGI', 'Impact_Category']

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Select features and target
X = df.drop(columns=['Impact_Category'])
y = df['Impact_Category']  # Now y will contain encoded numerical values

# Standardize numerical features
scaler = StandardScaler()

# Handle class imbalance using SMOTE with more conservative settings
smote = SMOTE(random_state=42, k_neighbors=5, sampling_strategy='auto')
X_scaled = scaler.fit_transform(X)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Train Random Forest Classifier with adjusted parameters
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)

# Train XGBoost Classifier with adjusted parameters
xgb_model = XGBClassifier(
    eval_metric='mlogloss',
    max_depth=4,
    learning_rate=0.1,
    n_estimators=100,
    min_child_weight=2,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)

# Train SVM Classifier
svm_model = SVC(kernel='rbf', probability=True, random_state=42)
svm_model.fit(X_train, y_train)
svm_preds = svm_model.predict(X_test)

# Evaluate performance
def evaluate_model(model_name, y_true, y_pred):
    print(f"{model_name} Performance:")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Classification Report:")
    print(classification_report(y_true, y_pred, zero_division=1))  # Add zero_division parameter
    print("Confusion Matrix:")
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

# Display evaluation results
evaluate_model("Random Forest", y_test, rf_preds)
evaluate_model("XGBoost", y_test, xgb_preds)
evaluate_model("SVM", y_test, svm_preds)
```
